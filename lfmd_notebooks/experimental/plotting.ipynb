```python
x = np.linspace(0, np.pi, 20)
offsets = np.linspace(0, 2 * np.pi, 8, endpoint=False)
yy = [np.sin(x + phi) for phi in offsets]
```

```python
with mpl.rc_context({}):
    for y in yy:
        plt.plot(x, y, linewidth=3)
    plt.show()
```

```python
with mpl.rc_context({'axes.prop_cycle': cycler('color', tol_colors.tol_cset('muted'))}):
    for y in yy:
        plt.plot(x, y, linewidth=3)
    plt.show()
```

```python
with mpl.rc_context(tue_cycler(color=tol_colors.tol_cset('medium-contrast'))):
    for y in yy:
        plt.plot(x, y, linewidth=3)
    plt.show()
```
